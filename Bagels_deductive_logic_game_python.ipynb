```python
def main():
    
    print("")
    print("Bagels, a deductive logic game.")
    print("")
    print("I am thinking of a 3-digit number. Try to guess what it is. Here are some clues:")
    print("")
    print("When I say:""          ""That means:")
    print("  Pico""               ""One digit is correct but in the wrong position.")
    print("  Fermi""              ""One digit is correct and in the right position.")
    print("  Bagels""             ""No digit is correct." )
    print("")
    print("I have thought up a number.")
    print("")
    print("You have 10 guesses to get it.")
      
    
    
    from random import randint
    rad_str = str(randint(100, 999))
    print("random number", rad_str)
    #rad_str = '345'
    i=1
    
    while i<=10:
        print("Guess#",i)
        i=i+1
        gus_num = str(input(""))
        
        
        if len(gus_num) != 3:
            print("This is NOT a 3 digit number!")
            
        if gus_num == rad_str:
            print('You got it!')
        
        clues_num = []
        
        for j in range(len(gus_num)):
            if gus_num[j] == rad_str[j]:
                clues_num.append('Fermi')
            elif gus_num[j] in rad_str:
                clues_num.append('Pico')
        
        if len(clues_num) == 0:
            print('Bagels')
        else:
            clues_num.sort()
            print(' '.join(clues_num))
            
        if i == 10 or gus_num == rad_str:
            print("Do you want to play again? (yes or no)")
            val = input("")
            if val == "yes":
                return main()
            else: 
                i=11
                print("Thanks for playing!")
```
